```python
# Define functions

# Read a CSV file - works for most files 
def input_csv(path,filename):
    raw_data = pd.read_csv(path + filename, sep=',',header=None, usecols=[0,1,2,3,4,6,10])  # leave off 13 (IRC) for now
    raw_data.columns = ['Stuff','Sail number','Boat design','Boat name','Elapsed time','TCF','PHRF']   # leave off IRC for now
    raw_data = raw_data.set_index('Boat name')
    return raw_data

# Read the rum race CSV files
def input_csv_rum(path,filename):
    raw_data = pd.read_csv(path + filename, sep=',',header=None, usecols=[0,1,2,3,7,10])
    raw_data.columns = ['Stuff','Sail number','Boat design','Boat name','Elapsed time','RumTCF']
    return raw_data

# Convert elapsed time (hh:mm:ss) to seconds as an integer
def et2seconds(x):
    if x[0] != '<':
        hours = int(x.split(':')[0])
        minutes = int(x.split(':')[1])
        seconds = int(x.split(':')[2])
        return seconds+ minutes*60 + hours *60*60

# Extract division Stuff field
def get_division(x):
    return x.split(',')[0].split(' ')[1]

# Extract start time from Stuff field
def get_starttime(x):
    time_str = x.split(',')[1].split('start time')[1].replace(' ','')
    return datetime.time(int(time_str[0:2]),int(time_str[2:]))
```

```python
# Define a dictionary called racedata and read race data from CSV files into it
racedata={}
for f in filenames:
    racedata[f.split('.')[0]] = input_csv(path,f)
for f in filenames3:
    racedata[f.split('.')[0]] = input_csv_rum(path,f)
```

```python
def boat_showed_up(x,boatname):
    x = x.ix[:,boatname,:].T
    return len(x[x['Sail number'].map(lambda x: not isnan(x))])
```

```python
def boat_dnf_counter(x):
    x = x.ix[:,boatname,:].T
    return len(x[x['Elapsed time']=='< dnf >'])
```
