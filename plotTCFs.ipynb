```python
inputFile = 'AllTCFs.csv'
handicap  = 'PNR'

rotateThreshold = 9

# Handicap types in 2013-14 season:
# Championship, Fleet, Offshore, Two-handed, Rum, Classics, and Island Bay.

import pandas as pd
import datetime as dt
import dateutil
import eaplots.plots as eap
import eaplots.reportwriter as eaprw
import cStringIO
import math
import os

datapath  = os.getcwd()
outputDir = os.path.join(datapath, 'Plots', handicap)

# Make output folder if it doesn't already exist
if not os.path.exists(outputDir):
    os.makedirs(outputDir)

pd.set_option('display.expand_frame_repr', True)
pd.set_option('display.line_width',10000)
pd.set_option('display.max_columns',10000)
pd.set_option('display.max_rows',10000)
pd.set_option('display.max_colwidth',10000)
pd.set_option('display.max_info_columns',10000)
pd.set_option('display.height',1000)
pd.set_option('display.max_seq_items',100)
pd.set_option('display.multi_sparse', True)

todaysDate = dt.datetime.strftime(dt.datetime.now(), "%d %B %Y")
#reportTitle = "TCFs - " + handicap + ' (Combined)'
#reportTitle = "TCFs - " + handicap
reportTitle = "TCFs - Port Nicholson regatta"
subtitle = todaysDate
secondPageTitle = "Notes"
footerText = todaysDate
bullets = ["Applied TCF is the TCF that was used on race day to compute corrected times on club handicap",
           "Actual TCF is the value the Applied TCF should have been if the race scoring software was programmed \
            precisely according to the TCF update formula (any observed differences between Applied TCF and Actual \
            TCF are very minor)",
           "Sailed-to TCF is a measure of how a boat performed on the day relative to the reference or median \
            boat for the fleet",
           "Clamped TCF is the clamped sailed-to TCF that feeds into the formula used to adjust TCFs ahead of \
            the next race. The clamps for 2013-14 were +/- 3.0%",
           "In the plots below, there are some boats that have no TCF data on particular days. This indicates that \
            the boat raced but was not given a score, e.g. retired after finishing, not entered, etc."]
```

```python
# Read in the TCF data from a CSV file
TCFdata = pd.read_csv(os.path.join(datapath,inputFile),sep=',',parse_dates=True,dayfirst=True)

# Filter out all but the specified handicap type
TCFdata = TCFdata[TCFdata['Handicap']==handicap]
#TCFdata = TCFdata[TCFdata['Handicap']==handicap][TCFdata['Div']=='Combined']
#TCFdata = TCFdata[TCFdata['Handicap']==handicap][TCFdata['Div']=='Combined']

# Replace the nan's with zeroes
TCFdata.fillna(value=0)

# Get a unique list of boats from TCFdata and sort boats alphabetically
boats = list(set(TCFdata['Boat name']))
boats.sort()

# Rename some columns in TCFdata so that plot legends look prettier
TCFdata.rename(columns={'TCFapplied': 'Applied TCF', 'TCFactual': 'Actual TCF', \
                        'TCFst': 'Sailed-to TCF',    'TCFclmp': 'Clamped TCF'}, inplace=True)

# Figure out what the x-axis labels ought to be 
if handicap=='Championship' or handicap=='Fleet':
    TCFdata['Xlabels'] = TCFdata['Date'] + '\n' + TCFdata['Series'] + '\n' + TCFdata['Race']
else:
    TCFdata['Xlabels'] = TCFdata['Date'] + '\n' + TCFdata['Race']

# Select columns to pull out of TCFdata and put into boatData
boatcols = ['Date','PHRF','Applied TCF','Actual TCF','Sailed-to TCF','Clamped TCF','Series','Race','Xlabels']

TCFdata[boatcols].head(5)
```

```python
#Function for putting labels
def putLabelsOnOneLine(labels):
    import pandas as pd
    labels=boatData.Xlabels      
    labels = labels.values.tolist()
    for i,b in enumerate(labels):    
        labs = b.split('\n')                      
        if len(labs) == 3: #if there are 3 lines in the label        
            labels[i] = labs[0]+', '+labs[1]+' '+labs[2] 
    return pd.Series(labels)
```

```python
#Function for shortening labels
def shortenLabels(labels):
    import pandas as pd
    labels = labels.values.tolist()
    for i,b in enumerate(labels):    
        labs = b.split('\n')        
        if len(labs) == 3: #if there are 3 lines in the label
            series = labs[1] #get the series from the second line
            words = series.split(' ')
            if len(words) > 1:            
                words = words[0][0]+words[1][0]
            else:
                words = words[0][0:3]
            labels[i] = labs[0]+'\n'+words+'\n'+labs[2]    
    return pd.Series(labels)
```

```python
# Loop through boats and plot their PHRF, applied TCF, actual TCF, sailed-to TCF and clamped TCF

# First create a dict to collect image data for the PDF report
imgdata = {}
imgsize = {}


for boat in boats:
    
    ## I think the next 2 lines are causing a problem in cases where a boat has NANs assoicated with it
    ## For example, check out div 1, championship results for TCF data above - The Guarantee and Wedgetail don't
    ## appear in plots. and they have records with all NaNs 
    
    boatData = TCFdata[boatcols]
    boatData = boatData[TCFdata['Boat name']==boat]
    
    print boat
        
    columnsToPlot = boatData.columns.tolist()
    columnsToPlot.remove('Date')
    columnsToPlot.remove('Series')
    columnsToPlot.remove('Race')
    columnsToPlot.remove('Xlabels')
    if handicap == 'Rum':
        columnsToPlot.remove('PHRF')
        
    #Adjust x labels:
    xTickRotate = 0
    aspectRatio = 0.497
    boatData = boatData.reset_index()
    if len(boatData) >= rotateThreshold:
        boatData.Xlabels=putLabelsOnOneLine(boatData.Xlabels)
        xTickRotate = 90  
        aspectRatio = 0.609

    eap.plotLine(
        boatData,
        figstyle = 2,
        figsize = 4,
        datatype = 'd',
        outPath = outputDir,
        filename = boat,             
        fileformat = '.png',
        xlabel = '',
        ylabel = 'TCF',
        columnsToPlot = columnsToPlot,
        xTickLabels = 'Xlabels',
        title = boat + ' - ' + handicap,
        ylim = None,
        decPlaces = 3,
        xTickRotation = xTickRotate
        )        

    # Capture plot to go into PDF report   
    imgdata[boat] = cStringIO.StringIO()
    imgsize[boat] = aspectRatio   
    gcf().savefig(imgdata[boat], format = 'PDF',  bbox_inches = 'tight')
```

```python
eaprw.createReport(reportTitle = reportTitle,
                   subtitle = subtitle,
                   footerText = footerText,
                   secondPageTitle = secondPageTitle,
                   bullets = bullets,
#                   filename = 'TCF report - ' + handicap + ' - Combined', 
#                   filename = 'TCF report - ' + handicap,
                   filename = 'TCF report - PNR',
                   imgsize = imgsize,
                   imgdata = imgdata,
                   )
```
