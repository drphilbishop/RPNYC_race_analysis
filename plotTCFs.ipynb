```python
import pandas as pd
import datetime as dt
import dateutil
import eaplots.plots as plots

pd.set_option('display.expand_frame_repr', True)
pd.set_option('display.line_width',10000)
pd.set_option('display.max_columns',10000)
pd.set_option('display.max_rows',10000)
pd.set_option('display.max_colwidth',10000)
pd.set_option('display.max_info_columns',10000)
pd.set_option('display.height',1000)
pd.set_option('display.max_seq_items',100)
pd.set_option('display.multi_sparse', True)

datapath = 'C:\\Phil\\My Dropbox\\Python\\Sailing\\'
#datapath = '/Users/PhilBishop/Dropbox/Python/Data/'

filename = 'AllTCFs.csv'

# Read input CSV file
#TCFdata = pd.read_csv(datapath+filename,sep=',',parse_dates=True,dayfirst=True,index_col='Date')
TCFdata = pd.read_csv(datapath+filename,sep=',',parse_dates=True,dayfirst=True)
```

```python
TCFdata.head(5)
```

```python
boat = 'Rocket Science'

boatcols = ['Date','PHRF','TCFapplied','TCFactual','TCFst','TCFclmp','Series','Race']
boatData = TCFdata[boatcols]
boatData = boatData[TCFdata['Boat name']==boat][TCFdata['Handicap']=='Championship']

boatData
```

```python
boatData['DSR'] = boatData['Date'] + '\n' + boatData['Series'] + '\n' + boatData['Race']
columnsToPlot = boatData.columns.tolist()
columnsToPlot.remove('Date')
columnsToPlot.remove('Series')
columnsToPlot.remove('Race')
columnsToPlot.remove('DSR')

plots.plotLine(
    boatData,
    figstyle = 2,
    figsize = 4,
    datatype = 'd',
    filename = boat,             
    fileformat = '.png',
    xlabel = 'Race',
    ylabel = 'TCF',
    plotAllCols = 'no',
    columnsToPlot = columnsToPlot,
    xTickLabels = 'DSR',
    ylim = (0.78,0.87)
    )
```
