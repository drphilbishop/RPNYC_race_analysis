#### TCFanalysis - code to read in a CSV file of RPNYC race results and check that the TCF updating procedure has been applied correctly.

```python
import pandas as pd
import datetime as dt
from functions import *

pd.set_option('display.expand_frame_repr', True)
pd.set_option('display.line_width',10000)
pd.set_option('display.max_columns',10000)
pd.set_option('display.max_rows',10000)
pd.set_option('display.max_colwidth',10000)
pd.set_option('display.max_info_columns',10000)
pd.set_option('display.height',1000)
pd.set_option('display.max_seq_items',100)
pd.set_option('display.multi_sparse', True)

datapath = 'C:\\Phil\\My Dropbox\\Python\\Sailing\\Data\\'
#datapath = '/Users/PhilBishop/Dropbox/Python/Sailing/Data/'

filename = 'All_results2013-14.csv'

# Set parameter values for TCF update formula
numSeeds = 2.
numFinished = 4.
pctFinished = 0.5
loClamp = .03
upClamp = .03    
```

```python
# Read input CSV file
df = pd.read_csv(datapath+filename,sep=',',parse_dates=True,dayfirst=True)

# Convert date format from dd/mm/yyyy to yyyy-mm-dd
df['Date'] = df['Date'].map(lambda x: dt.date(day=int(x.split('/')[0]), \
                                         month=int(x.split('/')[1]), \
                                         year=int(x.split('/')[2])))

# Initialise dicts keyed by a boatname-handicap tuple
newTCFvalue = {}                # dict for storing newTCF
TCFclampHistoryAndSeeds = {}    # dict for storing seed TCFs and clamped TCFs

# Loop through every row in the transposed dataframe and establish keyed lists
for index,row in df.T.iteritems():

    # Create a set of unique keys for boat-handicap tuples    
    key = row['Boat name'], row['Handicap']    

    # For each key value, create two empty lists
    newTCFvalue[key] = []   
    TCFclampHistoryAndSeeds[key] = []

# Index the dataframe by date, handicap, series, race and div, and sort it
df = df.set_index(['Date','Handicap','Series','Race','Div']).sort()

# Make a new dataframe for appending new TCF results to
dfTCF = pd.DataFrame()

# Iterate through the raw/original dataframe on each unique race, indexed on date-handicap-
# series-race-div. At each iterate, collect results into a dataframe called 'race'.
for index, race in df.groupby(level=(0,1,2,3,4)):   
    
    key = index[1]   
    
    # Compute TCFactual (what the TCF should've been on the day)
    race['TCFactual'] = race.apply(TCFactual,axis=1,indexKey=key,newTCFvalue=newTCFvalue)
    
    # Compute corrected time, corrSecs, using TCFactual
    race['corrSecs'] = race.apply(corrSecs,axis=1)
        
    # Compute rank based on corrected time and compare with rank applied on the day
    race['Rank'] = race['corrSecs'].rank(ascending=True)
    race['rankDiff'] = race.apply(rankDiff,axis=1)

    # Figure out the reference boat and reference time
    race['refTime'] = race.groupby(level=(0,1,2,3,4)).apply(ref_boat)
    
    # Replace reftime value with nan for boats that do not race 
    race['refTime'] = race.apply(adjustRefTime,axis=1)                
        
    # Compute TCFst, pctMvmt, TCFclmp
    race['TCFst']   = race.apply(TCFst,axis=1)
    race['pctMvmt'] = race.apply(pctMvmt,axis=1)
    race['TCFclmp'] = race.apply(lambda x: TCFclmp(x,loClamp=loClamp,upClamp=upClamp),axis=1)
    
    # If history of clamped TCFs is empty (coz boat has done no races), add seed values; Otherwise add TCFclmp
    race.apply(updateTCFclampAndSeed,axis=1,race=race,indexKey=key,TCFclampHistoryAndSeeds=TCFclampHistoryAndSeeds)  
    
    # Compute newTCF
    race['newTCF'] = race.apply(newTCF,axis=1,race=race,indexKey=key,TCFclampHistoryAndSeeds=TCFclampHistoryAndSeeds,newTCFvalue=newTCFvalue)
    
    # Compute the difference between TCFapplied and TCFactual
    race['TCFdiff'] = race.apply(TCFdiff,axis=1)

    # Sort race by rank (ascending)
    race = race.set_index(['Rank'],append=True).sort()
    
    # Now, append the race dataframe to the dfTCF dataframe
    dfTCF = dfTCF.append(race.reset_index())
    
# Set column order for dfTCF
cols=['Boat name','Elapsed time','ETsecs','corrSecs','refTime','RankAppld','Rank','rankDiff','PHRF','IRC', \
      'TCFapplied','TCFactual','TCFdiff','TCFst','pctMvmt','TCFclmp','newTCF']
    
# Reindex dfTCF, sorting by index
dfTCF = dfTCF.set_index(['Date','Handicap','Series','Race','Div']).reindex(columns=cols).sort().reset_index()

# Reset index, apply date-to-string conversion to get short month format, and then reindex without sorting
dfTCF.reset_index(inplace=True)
dfTCF['Date'] = dfTCF['Date'].map(lambda x: str(x.day) + ' ' + dt.datetime.strftime(x,'%b'))
dfTCF = dfTCF.set_index(['Date','Handicap','Series','Race','Div']).reindex(columns=cols)
```
