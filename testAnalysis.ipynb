```python
# Compute corrected PHRF seconds
def secsPHRF(x):
    if '<' in x['Elapsed time']:
        return np.nan
    else:
        return x['PHRF'] * x['secsET']

data['secsPHRF'] = data.apply(secsPHRF, axis=1)
#data.head(20)
```
