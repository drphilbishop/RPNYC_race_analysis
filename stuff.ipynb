```python
# Define functions

# Read selected columns from raw CSV file and make boat name the index
def input_csv(path,filename):
    raw_data = pd.read_csv(path + filename, sep=',', header=None, usecols=CSVcolNums)
    raw_data.columns = CSVcolHdrs
    raw_data = raw_data.set_index('Boat name')
    return raw_data

# Convert elapsed time (hh:mm:ss or mm:ss or <xxx>) to seconds as an integer
def et2seconds(x):
    if '<' in x:
        return np.nan    # preferable to 'NaN', NaN, or 0
    else:
        if len(x) > 5:
            hours = int(x.split(':')[0])
            minutes = int(x.split(':')[1])
            seconds = int(x.split(':')[2])
            return seconds + minutes*60 + hours*60*60
        else:
            minutes = int(x.split(':')[0])
            seconds = int(x.split(':')[1])
            return seconds + minutes*60
            
# Extract division from Stuff field
def get_division(x):
    if 'Division' in x:
        return x.split('Division')[1].replace(' ','')[0]  # assumes returned value is a single character - will fail if div is not, 1, 2, 3, A, B, etc
    else:
        return 'Combined'

# Extract start time from Stuff field
def get_starttime(x):
#   time_str = x.split(',')[1].replace(' ','')[-4:]
    time_str = x.split('ime ')[1][:4]
    return datetime.time(int(time_str[0:2]),int(time_str[2:])) # time_str yields hhmm - try to display as hh:mm c.f. hh:mm:ss

# Determine the reference boat, i.e. boat with median corrected time or nearest boat quicker than the median corrected time
def ref_boat(df):
    x = df.corSecs.dropna()    # x is the series formed by extracting the column called corSecs from df
    m = x.median()             # m is the median value from x  
    if (x==m).sum()==0:        # (x==m) returns a boolean; the sum equals zero if no boat's time is the median
        refboat = x[x < m].tail(1).index[0]  # grab the last boat from boats above the median
    else:
        refboat = x[x==m].index[0]
    return refboat, m

# Compute the sailed to handicap for this race
def computeST(x):
    if x['corSecs'] == NaN:
        return NaN
    else:
        return round( (x['TCF'] * reftime / x['corSecs']), 3)

# Compute perentage movement in TCF, i.e. ST versus TCF    
def pctmovement(x):
    if x['ST'] == NaN:
        return NaN
    else:
        return round(100 * (x['ST'] - x['TCF']) / x['TCF'], 2)

# Compute clamped sailed to handicap 
def clampST(x):
    #print x
    # case where boat is the reference boat or the boat was never scored
    if (x['corSecs'] == reftime) or (x['corSecs'] == 'Nan'):
        return x['TCF']
    # case where boat is better than reference boat
    if x['corSecs'] < reftime:
        if abs(x['ST'] - x['TCF']) / x['TCF'] > upClamp:
           return round(((1 + upClamp) * x['TCF']), 3)
        else:
           return x['ST']
    # case where boat is worse than reference boat
    if x['corSecs'] > reftime:
        if abs(x['ST'] - x['TCF']) / x['TCF'] > loClamp:
           return round(((1 - loClamp) * x['TCF']), 3)
        else:
           return x['ST']
```

```python
# Create a uniformly structured dataframe of current race data

data = q
data = data.replace(0.00,nan)

data['Race no'] = 5
data['Date']    = "Saturday, 16 February 2013"
data['Type']    = "Divisional"
data['Series name'] = "New Year" 

data['ID'] = filename.split('id')[1] 
data['Division']   = data['Stuff'].map(lambda x: get_division(x))
data['Start time'] = data['Stuff'].map(lambda x: get_starttime(x))
data['Seconds']    = data['Elapsed time'].map(lambda x: et2seconds(x))
data['corSecs']    = data['Seconds'] * data['TCF']
data
```
