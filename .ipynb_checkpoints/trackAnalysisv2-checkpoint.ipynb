```python
import pandas as pd
import numpy as np
import datetime as dt
#import functions as fn
#import matplotlib
import os
import re
import xml.etree.ElementTree as et

# Variables
filename = 'test1.vcc'
start_time = '2015-03-21 14:00:00' 
finish_time = '2015-03-21 15:51:37'
angle = 20

pd.set_option('display.width',10000)
pd.set_option('display.precision', 4)
pd.set_option('display.max_columns',10000)
pd.set_option('display.max_rows',10000)
pd.set_option('display.max_colwidth',10000)
pd.set_option('display.max_info_columns',10000)
pd.set_option('display.notebook_repr_html',True)
pd.set_option('display.expand_frame_repr', True)
```

```python
cd 'C:\Users\bishopp\Dropbox\Python\Sailing'
```

```python
# Parse XML file and extract Trackpoints node
tree = et.parse(filename)
root = tree.getroot()
trackpoints = root.getchildren()[0].getchildren()[5].getchildren()
elems = [(tp.attrib['dateTime'],tp.attrib['heading'],tp.attrib['speed'],tp.attrib['latitude'],tp.attrib['longitude'])
        for tp in trackpoints]
```

```python
df = pd.DataFrame(elems, columns=['Time','Heading','Speed','Lat','Long'])
df['Time'] = df['Time'].apply(lambda x: dt.datetime.strptime(x, '%Y-%m-%dT%H:%M:%S+13:00'))
df['Heading'] = df['Heading'].apply(float)
df = df[df['Time']>=start_time][df['Time']<=finish_time]
df['Change'] = df['Heading'].diff().apply(lambda x: 360-abs(x) if abs(x)>180 else abs(x)).fillna(0)
df.head()
```

```python
df[df['Change'] > angle]
```

```python
df.to_csv('changeinTracks.csv', sep = ',',index = False)
```
