```python
# Define functions

# Read selected columns from raw CSV file and make boat name the index
def input_csv(path,filename):
    raw_data = pd.read_csv(path + filename, sep=',', header=None, usecols=CSVcolNums)
    raw_data.columns = CSVcolHdrs
    raw_data = raw_data.set_index('Boat name')
    return raw_data

# Convert elapsed time (hh:mm:ss or mm:ss or <xxx>) to seconds as an integer
def et2seconds(x):
    if '<' in x:
        return np.nan    # preferable to 'NaN', NaN, or 0
    else:
        if len(x) > 5:
            hours = int(x.split(':')[0])
            minutes = int(x.split(':')[1])
            seconds = int(x.split(':')[2])
            return seconds + minutes*60 + hours*60*60
        else:
            minutes = int(x.split(':')[0])
            seconds = int(x.split(':')[1])
            return seconds + minutes*60
            
# Extract race number from Race field
def get_race(x):
    if 'ace ' in x:
        return x.split('ace ')[1].replace(' ','-').split('-')[0]
    else:
        return NaN

# Extract division from Stuff field
def get_division(x):
    if 'Division' in x:
        return x.split('Division')[1].replace(' ','')[0]  # assumes returned value is a single character - will fail if div is not, 1, 2, 3, A, B, etc
    else:
        return 'Combined'

# Extract start time from Stuff field
def get_starttime(x):
#   time_str = x.split(',')[1].replace(' ','')[-4:]
    time_str = x.split('ime ')[1][:4]
    return datetime.time(int(time_str[0:2]),int(time_str[2:])) # time_str yields hhmm - try to display as hh:mm c.f. hh:mm:ss

# Extract course number from Stuff field
def get_course(x):
#   return x.split(',')[2].split(' ')[2]
    if 'course' in x:
        return x.split('course')[1].replace(' ','').split(',')[0] # assumes course number is followed by a comma, which it would if distance is there too
    else:
        return NaN

# Extract course distance (nm) from Stuff field
def get_miles(x):
#   return x.split(',')[3].split(' ')[2]
    if 'distance' in x:
        return x.split('distance')[1].replace(' ','-').split('-')[1] # assumes value to be returned is a single character - will fail if div is not, 1, 2, 3, A, B, etc
    else:
        return NaN

# Determine the reference boat, i.e. boat with median corrected time or nearest boat quicker than the median corrected time
def ref_boat(df):
    x = df.corSecs.dropna()    # x is the series formed by extracting the column called corSecs from df
    m = x.median()             # m is the median value from x  
    if (x==m).sum()==0:        # (x==m) returns a boolean; the sum equals zero if no boat's time is the median
        refboat = x[x < m].tail(1).index[0]  # grab the last boat from boats above the median
    else:
        refboat = x[x==m].index[0]
    return refboat, m

# Compute the sailed to handicap for this race
def computeST(x):
    if x['corSecs'] == NaN:
        return NaN
    else:
        return round( (x['TCF'] * reftime / x['corSecs']), 3)

# Compute perentage movement in TCF, i.e. ST versus TCF    
def pctmovement(x):
    if x['ST'] == NaN:
        return NaN
    else:
        return round(100 * (x['ST'] - x['TCF']) / x['TCF'], 2)

# Compute clamped sailed to handicap 
def clampST(x):
    #print x
    # case where boat is the reference boat or the boat was never scored
    if (x['corSecs'] == reftime) or (x['corSecs'] == 'Nan'):
        return x['TCF']
    # case where boat is better than reference boat
    if x['corSecs'] < reftime:
        if abs(x['ST'] - x['TCF']) / x['TCF'] > upClamp:
           return round(((1 + upClamp) * x['TCF']), 3)
        else:
           return x['ST']
    # case where boat is worse than reference boat
    if x['corSecs'] > reftime:
        if abs(x['ST'] - x['TCF']) / x['TCF'] > loClamp:
           return round(((1 - loClamp) * x['TCF']), 3)
        else:
           return x['ST']
```

```python
# Take a quick look at imported raw CSV file - need to select/deselect appropriate cols in CSV

# Winter series fully-crewed raw CSV column numbers and headers
CSVcolNums = [1,3,5,7,9,28,29,30,31,32,35,41,45]
CSVcolHdrs = ['Race','Date','Wind direction','Wind speed','ID','Stuff','Sail number','Boat design', \
              'Boat name','Elapsed time','TCF','PHRF','IRC']

# Winter series 2-handed raw CSV column numbers and headers
#CSVcolNums = [1,3,5,7,9,24,25,26,27,28,31,37]
#CSVcolHdrs = ['Race','Date','ID','Wind direction','Wind speed','Stuff','Sail number','Boat design', \
#              'Boat name','Elapsed time', 'PHRF','IRC']

# Have a quick look at a single raw CSV file
filename = 'id13 DivB r5'
q = input_csv(datapath + "Raw CSV data/", filename + '.csv')
#q.head(5)
q

# id1 DivA r1, id2 DivA r2, id5 DivA r3,  id9 DivA r4,  id12 DivA r5   # div A 2013 winter series
# id3 DivB r1, id6 DivB r2, id10 DivB r3, id11 DivB r4, id13 DivB r5   # div B 2013 winter series
# id4 2h r1,   id7 2h r2,   id8 2h r3     # 2-handed 2013 winter series
```

```python
# Create a uniformly structured dataframe of current race data

# First eyeball raw file above and specify the following string variables 
seriesType = "Winter"      # Championship, Twilight, Two-handed, Rum, At home, Offshore, Special, Winter, Classics, Regatta
seriesRace = "Winter"      # May be the same as type, o/wise Autumn, Spring, Anzac, etc
raceType   = "Fleet/Div"   # Divisional, Fleet/Cmbnd, Fleet/Div
windDir    = ""
windSpeed  = ""
comment    = "-"
data = input_csv(datapath + "Raw CSV data/", filename + '.csv')
data = data.replace(0.00,nan)
data['Series type'] = seriesType
data['Series name'] = seriesRace
data['Type']        = raceType
data['Comment']     = comment
#data['Wind direction'] = windDir     # Enter wind direction if not already there
#data['Wind speed']     = windSpeed   # Enter wind speed if not already there
data['Race no']     = data['Race'].map(lambda x: get_race(x))
data['Division']    = data['Stuff'].map(lambda x: get_division(x))
data['Start time']  = data['Stuff'].map(lambda x: get_starttime(x))
data['Course']      = data['Stuff'].map(lambda x: get_course(x))
data['Distance']    = data['Stuff'].map(lambda x: get_miles(x))
data['Seconds']     = data['Elapsed time'].map(lambda x: et2seconds(x))
data['corSecs']     = data['Seconds'] * data['TCF']    # multiply elapsed time in seconds by handicap used to score/rank this race - TCF, PHRF, etc
#data['corSecs']    = data['Seconds'] * data['PHRF']   # 2h
#data['TCF']        = 'NaN'                            # 2h

# Set upper and lower clamps on sailed to TCF
upClamp = .025
loClamp = .030

# Determine the reference boat and the reference corrected time
refboat = ref_boat(data)
reftime = data.ix[refboat[0]]['corSecs']

data['ST']      = data.apply(computeST, axis=1)
data['pctMvmt'] = data.apply(pctmovement, axis=1)
data['clpST']   = data.apply(clampST, axis=1)

# Identify desired columns in order, assign to resultsdata and sort by TCFsecs (also pops those cols not identified)
#data.columns.tolist()
cols = ['ID', 'Date', 'Series type', 'Series name', 'Type', 'Division', 'Race no', 'Start time', 'Course', 'Distance', 'Wind direction', 'Wind speed', 'Elapsed time', \
        'Seconds', 'corSecs', 'TCF', 'ST', 'pctMvmt', 'clpST', 'PHRF', 'IRC', 'Comment']

# cols2 is a shorter list than cols - easier to display and check the dataframe
cols2 = ['ID', 'Date', 'Series type', 'Series name', 'Type', 'Division', 'Race no', 'Elapsed time', 'Seconds', 'corSecs', 'TCF', 'ST', 'pctMvmt', 'clpST', 'PHRF']

# Sort on corrected time
resultsdata = data[cols2].sort('corSecs')
resultsdata
```
