```python
# Histograms

# Define function to generate bins - returns a list
def generateBins(start=None,interval=None,bins=None):
    binList = []; binList.append(start)
    for i in arange(1,bins+1):
        binList.append(start + i*interval)
    return binList   

# Histogram for the whole season (ie TCFdata)
#eap.plotHist(TCFdata, columnToPlot='Boat name', numBins=generateBins(start=0,interval=2,bins=10), figstyle=2)
eap.plotHist(TCFdata, columnToPlot='Boat name', numBins=10, figstyle=2)

ax = plt.gca()
ymajorLocator = tkr.MultipleLocator(base=1)
ax.yaxis.set_major_locator(ymajorLocator)

# Get list of handicaps existing in the dataframe
handicaps = TCFdata['Handicap'] ;
handicaps = list(handicaps.drop_duplicates())
print handicaps
```

```python
# Define a function for formatting x-axis labels
def putLabelsOnOneLine(labels):
    import pandas as pd
    labels=boatData.Xlabels      
    labels = labels.values.tolist()
    for i,b in enumerate(labels):    
        labs = b.split('\n')                      
        if len(labs) == 3: #if there are 3 lines in the label        
            labels[i] = labs[0]+', '+labs[1]+' '+labs[2] 
    return pd.Series(labels)

# Define a function for shortening labels
def shortenLabels(labels):
    import pandas as pd
    labels = labels.values.tolist()
    for i,b in enumerate(labels):    
        labs = b.split('\n')        
        if len(labs) == 3: #if there are 3 lines in the label
            series = labs[1] #get the series from the second line
            words = series.split(' ')
            if len(words) > 1:            
                words = words[0][0]+words[1][0]
            else:
                words = words[0][0:3]
            labels[i] = labs[0]+'\n'+words+'\n'+labs[2]    
    return pd.Series(labels)

def padYMax(plt,gca,padY):
    addToY = padY * getYTickInterval(plt,gca) # Amount to add to the existing y-max limit, where padY is the pad factor
    ymin,ymax = plt.ylim()
    plt.ylim(ymin,ymax+addToY)    

def getYTickInterval(plt,gca):
    yTickInterval=None
    ymin,ymax=plt.ylim()    
    for i,ticklabel in enumerate(gca.get_yticklabels()):      
        if len(gca.get_yticklabels()) > 1:
            if i==1:      
                value1 = ymin
                value2 = float(ticklabel.get_text())
                return value2 - value1        
```

```python
# Loop through boats and plot their PHRF, applied TCF, actual TCF, sailed-to TCF and clamped TCF

# First create a dict to collect image data for the PDF report
imgdata = {}
imgsize = {}

for boat in boats:
    
    boatData = TCFdata[boatcols]
    boatData = boatData[TCFdata['Boat name']==boat]
    
    print boat
        
    columnsToPlot = boatData.columns.tolist()
    columnsToPlot.remove('Date')
    columnsToPlot.remove('Series')
    columnsToPlot.remove('Race')
    columnsToPlot.remove('Xlabels')
    if handicapFilter == 'Rum':
        columnsToPlot.remove('PHRF')
        
    # Adjust x labels:
    xTickRotate = 0
    aspectRatio = 0.497
    boatData = boatData.reset_index()
    if len(boatData) >= rotateThreshold:
        boatData.Xlabels = putLabelsOnOneLine(boatData.Xlabels)
        xTickRotate = 90  
        aspectRatio = 0.609

    eap.plotLine(
        boatData,
        figstyle = 2,
        figsize = 4,
        datatype = 'd',
        outPath = outputDir,
        filename = boat,             
        fileformat = '.png',
        xlabel = '',
        ylabel = 'TCF',
        columnsToPlot = columnsToPlot,
        xTickLabels = 'Xlabels',
        title = boat,
        ylim = None,
        decPlaces = 3,
        xTickRotation = xTickRotate,
        forceLegendFrame = True
        )        
    
    # Pad the upper y-axis 
    yMaxPadding = 0.25   # add 25% of a tick interval at the y-axis max 
 
    gca = plt.gca() # get current axes  
    
    if yMaxPadding:
        padYMax(plt,gca,yMaxPadding) 

    # Capture plot to go into PDF report   
    imgdata[boat] = cStringIO.StringIO()
    imgsize[boat] = aspectRatio   
    gcf().savefig(imgdata[boat], format = 'PDF',  bbox_inches = 'tight')
```
