```python
# Take a quick look at a raw CSV file - need to select/deselect appropriate cols in CSV

infilename = 'id3 spring r3 results'

# Following list of colNums and colHdrs works fine on:
# - Divisional races
# - Winter series fully crewed
CSVcolNums = [1,3,5,7,9,28,29,30,31,32,35,41,45]
CSVcolHdrs = ['Race','Date','Wind direction','Wind speed','ID','Stuff','Sail number','Boat design', \
              'Boat name','Elapsed time','TCF','PHRF','IRC']

q = fn.input_csv(datapath + "Raw CSV data/", infilename + '.csv', CSVcolNums, CSVcolHdrs)
q.head(5)
#q

# Notes:
# - Edit '..\Handicapping\2013-14 season\Season summary (2013-14).xlsx'
# - Compare imported data with hard-copy printout from Yacht Systems
# - Make sure Elapsed time field begins with < if no time, e.g. <n/e>, <no entry>, <dnf>, <raf>, <osc>, <dns> etc
# - Check for duplicate boat names, i.e. incorrect spelling
```

```python
# Create a uniformly structured dataframe for the current race data

# Specify i for id, s for series name, r for race number and d for date - for the current set of results
i = "id3" 
s = "Spring"
r = "r3"
d = "5 Oct"

loClamp = .03    # Lower clamp on amount TCF can move in one day
upClamp = .03    # Upper clamp on amount TCF can move in one day

   
data = fn.input_csv(datapath + "Raw CSV data/", infilename + '.csv', CSVcolNums, CSVcolHdrs)
data = data.replace(0.00,nan)
data['secsET']  = data['Elapsed time'].map(lambda x: fn.et2seconds(x))
data['secsCor'] = data.apply(fn.secsCor, axis=1)
data['Div']     = data['Stuff'].map(lambda x: fn.get_division(x))
data['Details'] = i + ' ' + s + ' ' + r + ' ' + d
data['Race'] = s + ' ' + r

# Figure out reference boat and times and put ref tuples into a dataframe called reference
ref = data.groupby(['Div']).apply(lambda x: fn.ref_boat(x))
reference = pd.DataFrame({'refBoat':ref.map(lambda x: x[0]),'refTime':ref.map(lambda x: x[1])})

data['refBoat'] = data['Div'].map(lambda x: reference.ix[x,'refBoat'])
data['refTime'] = data['Div'].map(lambda x: reference.ix[x,'refTime'])

data['TCFst']   = data.apply(fn.TCFst, axis=1)
data['pctMvmt'] = data.apply(fn.pctMvmt, axis=1)
data['TCFclmp'] = data.apply(lambda x: fn.TCFclmp(x,loClamp=loClamp, upClamp=upClamp), axis=1)

# Identify desired columns - in order.
cols = ['Div', 'Race', 'Details', 'Elapsed time', 'secsET', 'secsCor', 'refTime', \
        'TCF', 'TCFst', 'pctMvmt', 'TCFclmp', 'PHRF', 'IRC']
#data.columns.tolist()

outfilename = 'All divs ' + s + ' ' + r
data[cols].to_csv(datapath + "Proc'd CSVs/" + outfilename + '.csv', sep = ',')

#data[cols].dtypes
data[cols].sort(['Div','secsCor'])
data[cols]
```

```python
# Count boats by division
data[cols].groupby(['Div']).count()
```

```python
# Create a separate dataframe for each division and write it to a CSV file

# Edit f according to divisions present in col Div above.
#f = ['div1', 'div2', 'div3', 'divClassics','divOpen']
f = ['div1', 'div2', 'div3', 'divOpen']

for x in f:
    div = x.split('div')[1]
    outfilename = i + ' ' + x + ' ' + s + ' ' + r
    x = data[cols][data['Div'] == div]
    x.to_csv(datapath + "Proc'd CSVs/" + outfilename + '.csv', sep = ',')
    print '\n', div, '  ', outfilename
    print x
```
