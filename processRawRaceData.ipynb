```python
# Define functions - if these functions are used frequently, consider moving them into a package.

# Read selected columns from raw CSV file and make boat name the index
def input_csv(path,filename):
    raw_data = pd.read_csv(path + filename, sep=',', header=None, usecols=CSVcolNums)
    raw_data.columns = CSVcolHdrs
    raw_data = raw_data.set_index('Boat name')
    return raw_data


# Convert elapsed time (hh:mm:ss or mm:ss or <xxx>) to seconds as an integer
def et2seconds(x):
    if '<' in x:
        return np.nan    # np.nan is preferable to 'NaN', NaN, or 0
    else:
        if len(x) > 5:
            hours = int(x.split(':')[0])
            minutes = int(x.split(':')[1])
            seconds = int(x.split(':')[2])
            return seconds + minutes*60 + hours*60*60
        else:
            minutes = int(x.split(':')[0])
            seconds = int(x.split(':')[1])
            return seconds + minutes*60


# Compute corrected seconds
def secsCor(x):
    if x['TCF'] == NaN:
        return NaN
    else:
        return int(x['TCF'] * x['secsET'])


# Extract division from Stuff field
def get_division(x):
    if 'Division' in x:
        return x.split('Division')[1].replace(' ','')[0]  # assumes returned value is a single character - will fail if div is not, 1, 2, 3, A, B, etc
    else:
        if 'Open, start' in x:
            return x.split(', start')[0]
        else:
            if 'Classics, start' in x:
                return x.split(', start')[0]
            else:
                return 'Combined'

            
# Determine the reference boat and time as per Yotsys, i.e. boat with median corrected time or nearest boat
# quicker than the median corrected time
def ref_boat(df):
    x = df.secsCor.dropna()                   # x is a series formed by extracting the secsCor column from df
    x.sort()                                  # Ensure series x is sorted in order of ascending corrected seconds, secsCor
    m = x.median()                            # m is the median value from x  
    if (x==m).sum()==0:                       # (x==m) returns a boolean; the sum equals zero if no boat's time is the median
        refboat = x[x < m].tail(1).index[0]   # Let the last boat from those boats above the median be the reference boat
        reftime = int(x[x < m].tail(1).sum()) # Let the corrected time of the reference boat be the reference time 
    else:
        refboat = x[x==m].index[0]            # Let median boat be the reference boat
        reftime = m                           # Let the corrected time of the reference boat (also the median time) be the reference time
    return refboat, reftime


# Compute the sailed-to handicap for this race
def TCFst(x):
    if x['secsCor'] == NaN:
        return NaN
    else:
        return round( (float(x['TCF']) * float(x['refTime']) / float(x['secsCor'])), 3)


# Compute percentage movement in TCF, i.e. sailed to versus current TCF. Use my ST calc (TCFst) coz YotSys gets it wrong)    
def pctMvmt(x):
    if x['TCFst'] == 'NaN':
        return 'NaN'
    else:
        return round( float(100 * (x['TCFst'] - x['TCF']) / x['TCF']), 2)


# Compute clamped sailed-to handicap 
def TCFclmp(x):
    #print x
    # case where boat is the reference boat or the boat was never scored
    if (x['secsCor'] == x['refTime']) or (x['secsCor'] == 'Nan'):
        return x['TCF']
    # case where boat is better than reference boat
    if x['secsCor'] < x['refTime']:
        if abs(x['TCFst'] - x['TCF']) / x['TCF'] > upClamp:
           return round(((1 + upClamp) * x['TCF']), 3)
        else:
           return x['TCFst']
    # case where boat is worse than reference boat
    if x['secsCor'] > x['refTime']:
        if abs(x['TCFst'] - x['TCF']) / x['TCF'] > loClamp:
           return round(((1 - loClamp) * x['TCF']), 3)
        else:
           return x['TCFst']
   
```

```python
# Take a quick look at a raw CSV file - need to select/deselect appropriate cols in CSV

infilename = 'id3 spring r3 results'

# Following list of colNums and colHdrs works fine on:
# - Divisional races
# - Winter series fully crewed
CSVcolNums = [1,3,5,7,9,28,29,30,31,32,35,41,45]
CSVcolHdrs = ['Race','Date','Wind direction','Wind speed','ID','Stuff','Sail number','Boat design', \
              'Boat name','Elapsed time','TCF','PHRF','IRC']

q = input_csv(datapath + "Raw CSV data/", infilename + '.csv')
#q.head(5)
q

# Notes:
# - Edit '..\Handicapping\2013-14 season\Season summary (2013-14).xlsx'
# - Compare imported data with hard-copy printout from Yacht Systems
# - Make sure Elapsed time field begins with < if no time, e.g. <n/e>, <no entry>, <dnf>, <raf>, <osc>, <dns> etc
# - Check for duplicate boat names, i.e. incorrect spelling
```

```python
# Create a uniformly structured dataframe for the current race data

# Specify i for id, s for series name, r for race number and d for date - for the current set of results
i = "id3" 
s = "Spring"
r = "r3"
d = "5 Oct"

loClamp = .03    # Lower clamp on amount TCF can move in one day
upClamp = .03    # Upper clamp on amount TCF can move in one day

data = input_csv(datapath + "Raw CSV data/", infilename + '.csv')
data = data.replace(0.00,nan)
data['secsET']  = data['Elapsed time'].map(lambda x: et2seconds(x))
data['secsCor'] = data.apply(secsCor, axis=1)
data['Div']     = data['Stuff'].map(lambda x: get_division(x))
data['Details'] = i + ' ' + s + ' ' + r + ' ' + d
data['Race'] = s + ' ' + r

# Figure out reference boat and times and put ref tuples into a dataframe called reference
ref = data.groupby(['Div']).apply(lambda x: ref_boat(x))
reference = pd.DataFrame({'refBoat':ref.map(lambda x: x[0]),'refTime':ref.map(lambda x: x[1])})

data['refBoat'] = data['Div'].map(lambda x: reference.ix[x,'refBoat'])
data['refTime'] = data['Div'].map(lambda x: reference.ix[x,'refTime'])

data['TCFst']   = data.apply(TCFst, axis=1)
data['pctMvmt'] = data.apply(pctMvmt, axis=1)
data['TCFclmp'] = data.apply(TCFclmp, axis=1)

# Identify desired columns - in order.
cols = ['Div', 'Race', 'Details', 'Elapsed time', 'secsET', 'secsCor', 'refTime', \
        'TCF', 'TCFst', 'pctMvmt', 'TCFclmp', 'PHRF', 'IRC']
#data.columns.tolist()

outfilename = 'All divs ' + s + ' ' + r
data[cols].to_csv(datapath + "Proc'd CSVs/" + outfilename + '.csv', sep = ',')

#data[cols].dtypes
data[cols].sort(['Div','secsCor'])
data[cols]
```
