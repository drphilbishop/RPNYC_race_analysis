```python
def getTimeDifference(df):    
    if str(df['Finish']) == "nan":
        return ""
    else:    
       startTime = str(int(df['Start']))
       finishTime = str(int(df['Finish']))     
       startTimeDate = dt.datetime.strptime(startTime,'%H%M')
       finishTimeDate = dt.datetime.strptime(finishTime, '%H%M%S')
       timeDiff = finishTimeDate - startTimeDate   
       #timeDiff = timeDiff.seconds  # Use this to get seconds
       return timeDiff
```

```python
data['ElapsedTime'] = data.apply(getTimeDifference,axis=1)
data[['Boat name','Start','Finish','ElapsedTime']]
```
