```python
# Define functions - if these functions are used frequently, consider moving them into a package.

# Read selected columns from raw CSV file and make boat name the index
def input_csv(path,filename):
    raw_data = pd.read_csv(path + filename, sep=',', header=None, usecols=CSVcolNums)
    raw_data.columns = CSVcolHdrs
    raw_data = raw_data.set_index('Boat name')
    return raw_data


# Convert elapsed time (hh:mm:ss or mm:ss or <xxx>) to seconds as an integer
def et2seconds(x):
    if '<' in x:
        return np.nan    # np.nan is preferable to 'NaN', NaN, or 0
    else:
        if len(x) > 5:
            hours = int(x.split(':')[0])
            minutes = int(x.split(':')[1])
            seconds = int(x.split(':')[2])
            return seconds + minutes*60 + hours*60*60
        else:
            minutes = int(x.split(':')[0])
            seconds = int(x.split(':')[1])
            return seconds + minutes*60
            

# Extract division from Stuff field
def get_division(x):
    if 'Division' in x:
        return x.split('Division')[1].replace(' ','')[0]  # assumes returned value is a single char, will fail if div is not 1, 2, A, B, etc
    else:
        return 'Combined'


# Determine the reference boat and time, i.e. boat with median corrected time or nearest boat quicker than the median corrected time
def ref_boat(df):
    x = df.corSecs.dropna()    # x is the series formed by extracting the column called corSecs from df
    m = x.median()             # m is the median value from x  
    if (x==m).sum()==0:        # (x==m) returns a boolean; the sum equals zero if no boat's time is the median
        refboat = x[x < m].tail(1).index[0]  # grab the last boat from boats above the median
    else:
        refboat = x[x==m].index[0]
    return refboat, m

    
# Compute the sailed-to handicap for this race
def TCFst(x):
    if x['corSecs'] == NaN:
        return NaN
    else:
        return round( (float(x['curTCF']) * float(x['refTime']) / float(x['corSecs'])), 3)
    
    
# Record a 'Y' if TCFys = TCFst (i.e. do I get the same as YotSys), o/wise record a '-'
def stTCFsame(x):
    if (x['TCFys'] == 'NaN') or (x['TCFst'] == 'NaN'):
        return 'NaN'
    if round(x['TCFys'], 3) ==  round(x['TCFst'], 3):
        return 'Y'
    else:
        return 'n'


# Compute percentage movement in TCF, i.e. sailed to versus current TCF. Use my ST calc (TCFst) coz YotSys gets it wrong)    
def pctMvmt(x):
    if x['TCFst'] == 'NaN':
        return 'NaN'
    else:
        return round( float(100 * (x['TCFst'] - x['curTCF']) / x['curTCF']), 2)


# Compute clamped sailed-to handicap 
def TCFclmpd(x):
    #print x
    # case where boat is the reference boat or the boat was never scored
    if (x['corSecs'] == x['refTime']) or (x['corSecs'] == 'Nan'):
        return x['curTCF']
    # case where boat is better than reference boat
    if x['corSecs'] < x['refTime']:
        if abs(x['TCFst'] - x['curTCF']) / x['curTCF'] > upClamp:
           return round(((1 + upClamp) * x['curTCF']), 3)
        else:
           return x['TCFst']
    # case where boat is worse than reference boat
    if x['corSecs'] > x['refTime']:
        if abs(x['TCFst'] - x['curTCF']) / x['curTCF'] > loClamp:
           return round(((1 - loClamp) * x['curTCF']), 3)
        else:
           return x['TCFst']
```

```python
# Take a quick look at a raw CSV file of a performance report - need to select/deselect appropriate cols in CSV

infilename = 'id1 spring r1 perf'

# Following list of colNums and colHdrs works fine on:
# - Divisional races
# - Winter series fully crewed
CSVcolNums = [3,5,7,9,21,24,25,28,29]
CSVcolHdrs = ['Race','TCF type','Date','ID','Stuff','Boat name','corTime','curTCF','TCFys']

q = input_csv(datapath + "Raw CSV data/", infilename + '.csv')
#q.head(5)
q

# Notes:
# - Edit '..\Handicapping\2013-14 season\Season summary (2013-14).xlsx'
# - Compare imported data with hard-copy printout from Yacht Systems
# - Make sure Elapsed time field begins with < if no time, e.g. <n/e>, <no entry>, <dnf>, <raf>, <osc>, <dns> etc
# - Check for duplicate boat names, i.e. incorrect spelling
```

```python
# Create a uniformly structured dataframe of performance data for the current race

# Specify i for id, s for series name, r for race number and d for date - for the current set of results
i = "id1"      
s = "spring"
r = "r1"
d = "28 Sep"

loClamp = .03    # Lower clamp on amount TCF can move in one day
upClamp = .03    # Upper clamp on amount TCF can move in one day

data = input_csv(datapath + "Raw CSV data/", infilename + '.csv')
data = data.replace(0.00,nan)
data['corSecs'] = data['corTime'].map(lambda x: et2seconds(x))
data['Div']     = data['Stuff'].map(lambda x: get_division(x))
data['Details'] = i + ' ' + s + ' ' + r + ' ' + d

# Figure out reference boat and times and put ref tuples into a dataframe called reference
ref = data.groupby(['Div']).apply(lambda x: ref_boat(x))
reference = pd.DataFrame({'refBoat':ref.map(lambda x: x[0]),'refTime':ref.map(lambda x: x[1])})

data['refBoat'] = data['Div'].map(lambda x: reference.ix[x,'refBoat'])
data['refTime'] = data['Div'].map(lambda x: reference.ix[x,'refTime'])

data['TCFst']    = data.apply(TCFst, axis=1)
data['chkTCFst'] = data.apply(stTCFsame, axis=1)
data['pctMvmt']  = data.apply(pctMvmt, axis=1)
data['TCFclmpd'] = data.apply(TCFclmpd, axis=1)

# Identify desired columns in order
# sort on corrected time?
#data.columns.tolist()
cols = ['Details', 'Div','corTime','corSecs','refBoat','refTime','curTCF','TCFys','TCFst','chkTCFst', \
        'pctMvmt','TCFclmpd']

outfilename = i + ' ' + s + ' ' + r
data[cols].to_csv(datapath + "Proc'd CSVs/" + outfilename + ' perf.csv', sep = ',')

#data[cols].dtypes
#data[cols].sort('corSecs')  -- need to find a way to sort according to div and by corSecs to yield
# the same order as results reports from YotSys
data[cols]
```
