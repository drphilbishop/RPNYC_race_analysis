### computeTCFs - code to read in a CSV file of RPNYC race results and apply the TCF updating procedure.

```python
# Assign parameter values required by the TCF update formula
numSeeds = 2.
numFinished = 4.
pctFinished = 0.5
numPastPerfs = 4.
loClamp = .03
upClamp = .03    

inputDataFile = 'All_results2014-15.csv'
#inputDataFile = 'All_results2014-15_DivOpen.csv'
ostype = 'win'  # Enter mac or win

import pandas as pd
import datetime as dt
from functions import *
import os

pd.set_option('display.expand_frame_repr', True)
pd.set_option('display.width',10000)
pd.set_option('display.max_columns',10000)
pd.set_option('display.max_rows',10000)
pd.set_option('display.max_colwidth',10000)
pd.set_option('display.max_info_columns',10000)
pd.set_option('display.max_seq_items',100)
pd.set_option('display.multi_sparse', True)

if ostype == 'win':
    dataPathAndFile = os.getcwd() + '\\Data\\' + inputDataFile
else:
    dataPathAndFile = os.getcwd() + '/Data/' + inputDataFile

print dataPathAndFile
```

```python
# Read the input data file
df = Read_CSV(dataPathAndFile)

# Initialise dicts to be keyed by a boatname-handicap tuple
newTCFvalue = {}                # dict for storing newTCF
TCFclampHistoryAndSeeds = {}    # dict for storing seeded and clamped TCFs
currentAlgo = {}                # dict for storing current scoring algorithm type

# Loop through every row in the transposed dataframe and establish keyed lists (avoids KeyError)
for index,row in df.T.iteritems():
    # Create a set of unique keys for boat-handicap tuples    
    key = row['Boat name'], row['Handicap'], row['Div']
    # Create three empty lists indexed on key
    newTCFvalue[key] = []   
    TCFclampHistoryAndSeeds[key] = []
    currentAlgo[key] = []    

# Index the input data by date, handicap, series, race, and div, then sort it
df = df.set_index(['Date','Handicap','Series','Race','Div']).sort()

# Establish a new dataframe for collecting up TCF updates
dfTCF = pd.DataFrame()

df.head(50)
```

```python
# Iterate through the input data file (df) on each unique index value, i.e. date-handicap-
# series-race-div. At each iterate, apply the various functions and collect results into a
# dataframe called 'results'.
for index, results in df.groupby(level=(0,1,2,3,4)):   
    
    key = index[1]    
        
    # Compute TCFactual (what the TCF should've been on the day)
    results['TCFactual'] = results.apply(TCFactual,axis=1,indexKey=key,newTCFvalue=newTCFvalue)
    
    # Compute corrected time using TCFactual (1), PHRF (2) or IRC (3) - depends on value of Scoring
    results['corrSecs'] = results.apply(corrSecs,axis=1)
        
    # Compute rank based on corrected time and compare with rank applied on the day
    results['Rank'] = results['corrSecs'].rank(ascending=True)  
    results['rankDiff'] = results.apply(rankDiff,axis=1)
    
    # Compute PHRF and IRC rankings    
    scoring = results.Scoring.unique()[0]  # Get the unique scoring method for this race (either 1, 2, or 3)
    if scoring == 1 or scoring == 2:
        results['PHRF x ETsecs'] = results['PHRF'] * results['ETsecs']
        results['rankPHRF'] = results['PHRF x ETsecs'].rank(ascending=True)
    if scoring == 1 or scoring == 3:
        results['IRC x ETsecs'] = results['IRC'] * results['ETsecs'] 
        results['rankIRC'] = results['IRC x ETsecs'].rank(ascending=True)

    # Figure out the reference boat and reference time
    results['refTime'] = results.groupby(level=(0,1,2,3,4)).apply(ref_boat)
    
    # Replace reftime value with nan for boats that do not race 
    results['refTime'] = results.apply(adjustRefTime,axis=1)                
        
    # Compute TCFst, pctMvmt, TCFclmp
    results['TCFst']   = results.apply(TCFst,axis=1)
    results['pctMvmt'] = results.apply(pctMvmt,axis=1)
    results['TCFclmp'] = results.apply(lambda x: TCFclmp(x,loClamp=loClamp,upClamp=upClamp),axis=1)
    
    # Check if there is a change of scoring algorithm 
    results.apply(checkNewScoringAlgorithms, axis=1, race=results, indexKey=key, currentAlgo=currentAlgo)
    
    # If history of clamped TCFs is empty (coz boat has done no races or scoring algorithm has changed),
    # add seed values; otherwise add TCFclmp
    results.apply(updateTCFclampAndSeed,axis=1,race=results,indexKey=key,TCFclampHistoryAndSeeds=TCFclampHistoryAndSeeds)  
        
    # Compute newTCF        
    results['newTCF'] = results.apply(newTCF,axis=1,race=results,indexKey=key,TCFclampHistoryAndSeeds=TCFclampHistoryAndSeeds, \
                                      newTCFvalue=newTCFvalue)
    
    # Compute the difference between TCFapplied and TCFactual
    results['TCFdiff'] = results.apply(TCFdiff,axis=1)

    # If rank is NaN then fill with 0
    results['Rank'].fillna(0, inplace=True) # Only need fillna in a rank column when 0 finishes
    
    # Sort race by the appropriate column (ascending)
    if scoring == 1:        
        results = results.set_index(['Rank'],append=True).sort()
    elif scoring == 2:
        results['Rank'] = np.nan
        results = results.set_index(['rankPHRF'],append=True).sort()
    elif scoring == 3:
        results['Rank'] = np.nan
        results = results.set_index(['rankIRC'],append=True).sort()

    # Reset the index after sorting
    results = results.reset_index()

    # Now, append the race dataframe to the dfTCF dataframe
    dfTCF = dfTCF.append(results)
```

```python
# Set column order for dfTCF
cols=['Boat name','resetTCF','TCFalgo','Scoring','Elapsed time','ETsecs','corrSecs','refTime','RankAppld','Rank','rankDiff', \
      'PHRF','rankPHRF','IRC','rankIRC','TCFapplied','TCFactual','TCFdiff','TCFst','pctMvmt','TCFclmp','newTCF']
    
# Reindex dfTCF, sorting by index
dfTCF = dfTCF.set_index(['Date','Handicap','Series','Race','Div']).reindex(columns=cols).sort().reset_index()

# Reset index, apply date-to-string conversion to get short month format, and then reindex without sorting
dfTCF.reset_index(inplace=True)
dfTCF['Date'] = dfTCF['Date'].map(lambda x: str(x.day) + ' ' + dt.datetime.strftime(x,'%b'))
dfTCF = dfTCF.set_index(['Date','Handicap','Series','Race','Div']).reindex(columns=cols)

dfTCF.head()
```

```python
# Save new dataframe to csv
dfTCF[cols].to_csv('AllTCFs.csv', sep = ',')
```

```python
# Extract the most recently updated or used TCF/PHRF/IRC for each boat by handicap
lastTCF_df = pd.DataFrame()
dfTCF = dfTCF.reset_index()

for index, df in dfTCF.groupby(['Handicap','Boat name']):    
    df = df.dropna(subset=['newTCF']).tail(1)    
    df = df[['Date','Handicap','Boat name','newTCF','PHRF','IRC']]    
    lastTCF_df = lastTCF_df.append(df.reset_index(drop=True))      
lastTCF_df.reset_index(drop=True,inplace=True)

lastTCF_df.to_csv('MostRecentTCFs.csv', sep = ',')
lastTCF_df.head(12)
```
